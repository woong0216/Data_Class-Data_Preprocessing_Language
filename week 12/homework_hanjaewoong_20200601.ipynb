### 크롬드라이브 열기

```python
import pandas as pd
import requests
from selenium import webdriver
import time
import random
from tqdm import tqdm
import request
from bs4 import BeautifulSoup
driver = webdriver.Chrome('./data\\chromedriver.exe')
```

### url 설정

```python
url="https://news.daum.net/digital"
driver.get(url)
```

### 파싱하기

```python
resp = requests.get(url)
html_src = resp.text
soup = BeautifulSoup(html_src, 'html.parser')
```

### IT 여성&남성이 많이 본 링크 추출

```python
# 링크추출
links = []
for link in soup.find("ul", {"class":"list_ranking"}).findAll("a"):
    if 'href' in link.attrs:
        links.append(link.attrs['href'])
```

```python
links
```

### 각 링크를 통해 문장 추출

```python
contents = []
for link in links:
    resp = requests.get(link)
    html_src = resp.text
    soup = BeautifulSoup(html_src, 'html.parser')
    content = soup.find('div', {'id':'harmonyContainer'})     
    contents.append(content.getText().strip().replace('\n',' '))
```
